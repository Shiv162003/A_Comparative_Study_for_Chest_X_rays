```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage import io
import os

# Define the directories where your images are stored
normal_dir = 'C:\\Users\\ASUS\\Desktop\\uml project part 3\\Normal/'
tuberculosis_dir = 'C:\\Users\\ASUS\\Desktop\\uml project part 3\\Tuberculosis/'

# Define the directories where you want to save the output images
output_normal_dir = 'C:\\Users\\ASUS\\Desktop\\output Normal/'
output_tb_dir = 'C:\\Users\\ASUS\\Desktop\\output TB/'

# Load the images
normal_images = []
tuberculosis_images = []
for filename in os.listdir(normal_dir):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        img = io.imread(os.path.join(normal_dir, filename), as_gray=True)
        normal_images.append(img)
for filename in os.listdir(tuberculosis_dir):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        img = io.imread(os.path.join(tuberculosis_dir, filename), as_gray=True)
        tuberculosis_images.append(img)

# Combine the images into a single array
images = np.concatenate([normal_images, tuberculosis_images])

# Flatten the images into vectors
X_flat = images.reshape(images.shape[0], -1)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_flat)

# Plot the results
plt.scatter(X_pca[:len(normal_images), 0], X_pca[:len(normal_images), 1], label='Normal')
plt.scatter(X_pca[len(normal_images):, 0], X_pca[len(normal_images):, 1], label='Tuberculosis')
plt.legend()
plt.title('PCA on Chest X-ray Images')
plt.show()

# Save the output images
for i, img in enumerate(normal_images):
    io.imsave(os.path.join(output_normal_dir, f"normal_{i}.jpg"), img)

for i, img in enumerate(tuberculosis_images):
    io.imsave(os.path.join(output_tb_dir, f"tb_{i}.jpg"), img)
```
